```python
def predict(x, w, b):
    return w * x + b


def compute_mse_loss(X, y, w, b):
    total_loss = 0

    for x_i, y_i in zip(X, y):
        y_pred = predict(x_i, w, b)
        error = y_pred - y_i
        total_loss += error ** 2

    return total_loss / len(X)
```
